## importando as bibliotecas e fazendo requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
url_base = 'https://lista.mercadolivre.com.br/'
produto = input('Qual o produto voce deseja? ')

response = requests.get(url_base + produto)
print(url_base + produto)
```

```python
print(response.text)
```

```python
site = BeautifulSoup(response.text, 'html.parser')
print(site.prettify())
```
